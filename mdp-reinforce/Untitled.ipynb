```python
import csv
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
easy_steps = []
with open('easy_steps.txt', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in reader:
        easy_steps.append([int(el) for el in row])
easy_steps = np.log(np.asarray(easy_steps))
easy_times = []
with open('easy_time.txt', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in reader:
        easy_times.append([int(el) for el in row])
easy_times = np.log(np.asarray(easy_times))
iterations = np.arange(easy_steps.shape[1]) + 1
```

```python
print(iterations)
print(easy_steps[0])
print(easy_steps[1])
```

```python
plt.plot(iterations, easy_steps[0], label="VI")
plt.plot(iterations, easy_steps[1], label="PI")
plt.title("Small: Log steps vs max_iterations")
plt.legend()
# plt.show()
plt.savefig('small_step.png')
```

```python
plt.plot(iterations, easy_times[0], label="VI")
plt.plot(iterations, easy_times[1], label="PI")
plt.title("Small: Log time (ms) vs max_iterations")
plt.legend()
# plt.show()
plt.savefig('small_time.png')
```
